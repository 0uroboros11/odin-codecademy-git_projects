```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

life_exp_gdp = pd.read_csv('all_data.csv')
```

```python
print(life_exp_gdp.head())
```

```python
sns.lineplot(x = "Life expectancy at birth (years)", y = "GDP", hue = 'Country', data = life_exp_gdp, marker = '.')
plt.grid(True)
plt.xlabel('Life expectancy at birth (years)')
plt.ylabel('GDP')
plt.title("Correlation between GDP and life expectancy of a country")
```

```python
sns.lineplot(x = "Year", y = "GDP", hue = 'Country', data = life_exp_gdp, marker = '.')
plt.grid(True)
plt.xlabel('Year')
plt.xticks(rotation = 30)
plt.ylabel('GDP')
plt.title("GDP per year for 6 countries. \n Has GDP increased over time?")
```

```python
sns.lineplot(x = "Year", y = "Life expectancy at birth (years)", hue = 'Country', data = life_exp_gdp, marker = '.')
plt.grid(True)
plt.xlabel('Year')
plt.xticks(rotation = 30)
plt.ylabel('Life expectancy at birth (years)')
plt.title("Life expectancy at birth (years) for 6 countries. \n Has life expectancy increased over time?")
```

```python
avg_life_exp = life_exp_gdp.groupby('Country')["Life expectancy at birth (years)"].mean().reset_index()
sns.barplot(x = "Country", y = "Life expectancy at birth (years)", hue = 'Country', data = avg_life_exp)
plt.grid(True)
plt.xlabel('Country')
plt.xticks(rotation = 330, ha = 'left')
plt.ylabel('Life expectancy at birth (years)')
plt.title("Average life expectancy at birth (years) for 6 countries")
```
